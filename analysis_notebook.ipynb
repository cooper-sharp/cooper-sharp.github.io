## 3. Data Cleaning & Preparation

```python
# Convert numeric columns that are stored as strings
numeric_cols = ['demo_yrs', 'bisbas_reward_res', 'bisbas_drive', 'bisbas_fun_seek', 
                'bisbas_bis', 'susd_depression', 'susd_mania', 'spsrq_sr', 'spsrq_sp',
                'gsas_total', 'gsas_urge', 'gsas_thought']

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print("Converted numeric columns")
```

```python
# Create readable population labels
df['population'] = df['source'].map({
    'ad_data': 'Adolescent/Young Adult',
    'dvs_data': 'Older Adult',
    'g25_data': 'General Population'
})

# Create readable gender labels
df['gender_label'] = df['demo_gender'].map({
    '1': 'Male',
    '2': 'Female',
    '3': 'Other'
})

# Create age groups
df['age_group'] = pd.cut(df['demo_yrs'], 
                          bins=[0, 24, 34, 54, 100],
                          labels=['18-24', '25-34', '35-54', '55+'])

print("Created derived variables")
```

## 4. SQL Queries (Using SQLite)

You can write SQL queries to analyze the data.

```python
# Create in-memory SQLite database
conn = sqlite3.connect(':memory:')
df.to_sql('gambling_data', conn, index=False, if_exists='replace')

print("Data loaded into SQLite database")
```

```python
# Example SQL query: Gambling prevalence by source and gender
query = """
SELECT 
    source,
    demo_gender,
    COUNT(*) as total,
    SUM(CASE WHEN prob_gam = 'yes' THEN 1 ELSE 0 END) as with_problems,
    ROUND(100.0 * SUM(CASE WHEN prob_gam = 'yes' THEN 1 ELSE 0 END) / COUNT(*), 2) as prevalence_pct
FROM gambling_data
WHERE demo_gender IN ('1', '2')
  AND prob_gam IN ('yes', 'no')
GROUP BY source, demo_gender
ORDER BY source, prevalence_pct DESC
"""

pd.read_sql_query(query, conn)
```

## 5. Exploratory Visualizations

Create different types of visualizations to explore the data.

### 5.1 Age Distribution

```python
import plotly.graph_objects as go

# Create figure
fig = go.Figure()

# Add trace for each population
populations = ['All Studies'] + sorted(df['population'].dropna().unique().tolist())

for population in populations:
    if population == 'All Studies':
        data = df['demo_yrs'].dropna()
    else:
        data = df[df['population'] == population]['demo_yrs'].dropna()
    
    # Create histogram trace
    fig.add_trace(go.Histogram(
        x=data,
        nbinsx=20,
        name=population,
        marker=dict(
            color='#3498db',
            line=dict(color='#2c3e50', width=1.5)
        ),
        opacity=0.8,
        visible=(population == 'All Studies')  # Only 'All' visible by default
    ))

# Create dropdown buttons
buttons = []
for i, population in enumerate(populations):
    # Create visibility list (only show one trace at a time)
    visible = [False] * len(populations)
    visible[i] = True
    
    buttons.append(
        dict(
            label=population,
            method='update',
            args=[
                {'visible': visible}
            ]
        )
    )

# Add dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=buttons,
            direction='down',
            pad={'r': 10, 't': 10},
            showactive=True,
            x=0.02,
            xanchor='left',
            y=1.15,
            yanchor='top',
            bgcolor='white',
            bordercolor='#95a5a6',
            borderwidth=2,
            font=dict(size=12)
        )
    ],
    
    template='plotly',
    
    # Axes
    xaxis=dict(
        title='Age (years)',
        title_font={'size': 16, 'color': '#34495e'},
        gridcolor='#ecf0f1',
        showline=True,
        linewidth=2,
        linecolor='#95a5a6'
    ),
    yaxis=dict(
        title='Count',
        title_font={'size': 16, 'color': '#34495e'},
        gridcolor='#ecf0f1',
        showline=True,
        linewidth=2,
        linecolor='#95a5a6'
    ),
    
    # Plot area
    plot_bgcolor='white',
    paper_bgcolor='#f8f9fa',
    
    # Size
    height=600,
    width=600,
    
    # Margins
    margin=dict(l=50, r=50, t=80, b=50)
)

# Add annotation to explain dropdown
fig.add_annotation(
    text="Select study population:",
    xref="paper", yref="paper",
    x=0.02, y=1.18,
    showarrow=False,
    font=dict(size=11, color="#7f8c8d"),
    xanchor='left'
)

fig.show()
```

### 5.2 Gambling Prevalence

```python
# Calculate prevalence by population and gender
prevalence = df[df['prob_gam'].isin(['yes', 'no'])].groupby(['population', 'gender_label']).agg({
    'prob_gam': lambda x: (x == 'yes').sum() / len(x) * 100
}).reset_index()
prevalence.columns = ['Population', 'Gender', 'Prevalence %']

# Create bar chart
fig = px.bar(prevalence, 
             x='Population', 
             y='Prevalence %', 
             color='Gender',
             barmode='group',
             title='Gambling Problem Prevalence by Population and Gender',
             text='Prevalence %')

fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(template='plotly_white')
fig.show()
```

### 5.3 Mental Health Comparison

```python
# Compare mental health scores between gamblers and non-gamblers
mental_health_vars = ['susd_depression', 'susd_mania', 'bisbas_bis']

# Filter for valid data
mh_data = df[df['prob_gam'].isin(['yes', 'no'])].copy()

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, var in enumerate(mental_health_vars):
    data = mh_data.dropna(subset=[var, 'prob_gam'])
    
    # Box plot
    data.boxplot(column=var, by='prob_gam', ax=axes[i])
    axes[i].set_title(var.replace('_', ' ').title())
    axes[i].set_xlabel('Gambling Problem')
    axes[i].set_ylabel('Score')
    plt.sca(axes[i])
    plt.xticks([1, 2], ['No', 'Yes'])

plt.suptitle('Mental Health Scores: Gamblers vs Non-Gamblers', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Interactive violin plot
fig = go.Figure()

for gam_status in ['no', 'yes']:
    data = mh_data[mh_data['prob_gam'] == gam_status]['susd_depression'].dropna()
    fig.add_trace(go.Violin(
        y=data,
        name='Yes' if gam_status == 'yes' else 'No',
        box_visible=True,
        meanline_visible=True
    ))

fig.update_layout(
    title='Depression Scores Distribution',
    yaxis_title='Depression Score',
    xaxis_title='Has Gambling Problem',
    template='plotly_white'
)
fig.show()
```

### 5.4 Correlation Analysis

```python
# Select variables for correlation analysis
corr_vars = ['bisbas_reward_res', 'bisbas_drive', 'bisbas_fun_seek', 'bisbas_bis',
             'susd_depression', 'susd_mania', 'spsrq_sr', 'spsrq_sp']

# Calculate correlation matrix
corr_matrix = df[corr_vars].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Key Variables')
plt.tight_layout()
plt.show()
```

### 5.5 Scatter Plots & Relationships

```python
# Interactive scatter plot with Plotly
scatter_data = df.dropna(subset=['bisbas_reward_res', 'susd_depression', 'prob_gam'])

fig = px.scatter(scatter_data, 
                 x='bisbas_reward_res', 
                 y='susd_depression',
                 color='prob_gam',
                 trendline='ols',
                 title='Reward Sensitivity vs Depression',
                 labels={
                     'bisbas_reward_res': 'Reward Sensitivity',
                     'susd_depression': 'Depression Score',
                     'prob_gam': 'Gambling Problem'
                 },
                 opacity=0.6)

fig.update_layout(template='plotly_white')
fig.show()
```

### 5.6 Geographic Analysis

```python
# State-level prevalence
state_data = df[df['prob_gam'].isin(['yes', 'no'])].groupby('demo_state').agg({
    'prob_gam': [('total', 'count'), ('with_problems', lambda x: (x == 'yes').sum())]
}).reset_index()

state_data.columns = ['state', 'total', 'with_problems']
state_data['prevalence'] = state_data['with_problems'] / state_data['total'] * 100

# Filter for states with enough data
state_data = state_data[state_data['total'] >= 10].sort_values('prevalence', ascending=False)

# Top 15 states
top_states = state_data.head(15)

fig = px.bar(top_states, 
             x='prevalence', 
             y='state',
             orientation='h',
             title='Top 15 States by Gambling Problem Prevalence',
             labels={'prevalence': 'Prevalence (%)', 'state': 'State'},
             text='prevalence')

fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(template='plotly_white', height=600)
fig.show()
```

## 6. Advanced Analysis

### 6.1 Risk Factor Analysis

```python
# Compare risk factors across age groups
risk_data = df[df['prob_gam'].isin(['yes', 'no'])].copy()

# Group by age and gambling status
age_comparison = risk_data.groupby(['age_group', 'prob_gam']).agg({
    'bisbas_reward_res': 'mean',
    'bisbas_fun_seek': 'mean',
    'spsrq_sr': 'mean'
}).reset_index()

# Create subplot for each variable
fig = make_subplots(rows=1, cols=3, 
                    subplot_titles=('Reward Sensitivity', 'Fun Seeking', 'Sensation Seeking'))

variables = ['bisbas_reward_res', 'bisbas_fun_seek', 'spsrq_sr']

for i, var in enumerate(variables, 1):
    for gam_status in ['no', 'yes']:
        data = age_comparison[age_comparison['prob_gam'] == gam_status]
        fig.add_trace(
            go.Bar(x=data['age_group'], y=data[var],
                   name=f"Gambler: {'Yes' if gam_status == 'yes' else 'No'}",
                   showlegend=(i == 1)),
            row=1, col=i
        )

fig.update_layout(height=500, title_text="Risk Factors by Age Group", 
                  barmode='group', template='plotly_white')
fig.show()
```

### 6.2 Statistical Tests

```python
from scipy import stats

# T-test: Compare depression scores between gamblers and non-gamblers
gamblers = df[df['prob_gam'] == 'yes']['susd_depression'].dropna()
non_gamblers = df[df['prob_gam'] == 'no']['susd_depression'].dropna()

t_stat, p_value = stats.ttest_ind(gamblers, non_gamblers)

print("Depression Score Comparison:")
print(f"Gamblers: Mean = {gamblers.mean():.2f}, SD = {gamblers.std():.2f}, n = {len(gamblers)}")
print(f"Non-gamblers: Mean = {non_gamblers.mean():.2f}, SD = {non_gamblers.std():.2f}, n = {len(non_gamblers)}")
print(f"\nt-statistic = {t_stat:.3f}, p-value = {p_value:.4f}")

if p_value < 0.05:
    print("✓ Significant difference detected (p < 0.05)")
else:
    print("No significant difference (p >= 0.05)")
```
